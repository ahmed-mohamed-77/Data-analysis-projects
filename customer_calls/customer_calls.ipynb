```python
import pandas as pd
import re
import warnings

warnings.filterwarnings(action="ignore")
warnings.simplefilter(action="ignore", category=FutureWarning)
```

```python
df = pd.read_excel(r"D:\Python\data analysis projects\brest_canser\Customer Call List.xlsx")
df = df.set_index(keys="CustomerID")
print(f"DataFrame Shape: {df.shape}")
df.columns
```

```python
df["Last_Name"] = df["Last_Name"].apply(
    func=lambda x : re.sub(
        pattern=r"[/| |\._]+", repl="", string= x
    ) if isinstance(x, str) else "unknown"
)
```

```python
def formate_phone_number(number: str) -> str:
    pattern = r"((\d{3})[\||\-|/]?(\d{3})[\||\-|/]?(\d{4}))"
    if number == "None":
        number = number.replace("None", "unknown")
        return number 
    
    if match := re.search(pattern=pattern, string=number, flags=re.IGNORECASE):
        return f"{match.group(2)}-{match.group(3)}-{match.group(4)}"
df["Phone_Number"] = df["Phone_Number"].astype(dtype=str)
df["Phone_Number"] = df["Phone_Number"].apply(func=lambda x: formate_phone_number(number=x))
```

```python
df[["street_Address", "state", "zip_code"]] = df["Address"].str.split(pat=",", n=2, expand=True)
df.drop(labels="Not_Useful_Column", axis=1, inplace=True)
```

```python
df = pd.concat(objs=[df.iloc[:,:4], df.iloc[:,6:9], df.iloc[:,4:6]], axis=1)
```

```python
df["Paying Customer"].replace(to_replace=['Yes', 'No', 'N', 'Y'], value=[True, False, False, True], inplace=True)
df["Do_Not_Contact"].replace(to_replace=['No', 'N', 'Yes', 'Y'], value=[False, False, True, True], inplace=True)
df.rename(columns={"Paying Customer": "Paying_Customer"}, inplace=True)
```
